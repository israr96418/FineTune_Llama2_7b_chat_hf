```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer
import torch
import transformers
```

```python
model = "Israr-dawar/psychology_chatbot"
tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = transformers.pipeline(
    model = model,
    tokenizer = tokenizer,
    torch_dtype = torch.bfloat16,
    trust_remote_code = True,
    device_map = "auto",
    max_length = 1000,
    do_sample = True,
    top_k = 10,
    num_return_sequences = 1,
    eos_token_id=tokenizer.eos_token_id

)
```

```python
model_llm = HuggingFacePipeline(pipeline = pipeline , model_kwargs = {'temperature':0})
```

```python
from langchain import PromptTemplate, LLMChain
```

```python
import locale

locale.getpreferredencoding = lambda: "UTF-8"
# !pip install chainlit

```

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
from langchain.schema.output_parser import StrOutputParser


def should_finish(next_input):
  """Returns True if the next input indicates that the user wants to finish the conversation."""
  return next_input.lower() == "exit"

# Create a prompt template
template = """
     You are a good psychologist. Please share your thoughts on the following text:
     `{text}`
     Now, could you please ask a question related to this `{text}`?
"""
prompt = PromptTemplate(template=template, input_variables=["text"])
memory = ConversationBufferMemory()


# LLM chain with ConversationalBufferMemory object
chain = LLMChain(prompt=prompt, llm=model_llm, memory = memory, output_parser = StrOutputParser())

# Get the user's initial input

text = input("What else would you like to talk about? ")

# Run the model on the initial input
output = chain.run(text)

# Print the model's output
print(output)

print("memory: ", memory.load_memory_variables({}))

# Continue the conversation until the user decides to stop
while True:
    # Get the user's next input
    print("\n")
    next_input = input("User: ")


    if should_finish(next_input):
      break
    print("memory: ", memory.load_memory_variables({}))

    # Run the model on the next input , passing converstion history as input
    output = chain.run(next_input)
    print(output)

```

```python
%%writefile app.py

import streamlit as st
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer
from langchain import PromptTemplate, LLMChain
from pyngrok import ngrok
from langchain.schema.output_parser import StrOutputParser
import torch
import transformers
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

# model = "LangChain12/Final_psychologybot"
model = "Israr-dawar/psychology_chatbot"
tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map={"": 0},
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)

model_llm = HuggingFacePipeline(pipeline=pipeline, model_kwargs={'temperature': 0})

def should_finish(next_input):
    return next_input.lower() == "exit"

template = """
     You are a good psychologist. Please share your thoughts on the following text:
     `{text}`
     Now, could you please ask a question related to this `{text}`?
"""

prompt = PromptTemplate(template=template, input_variables=["text"])
memory = ConversationBufferMemory()

public_url = ngrok.connect(port='8501')  # Streamlit's default port is 8501

st.title("PsychologyBot")

# LLM chain with ConversationalBufferMemory object
chain = LLMChain(prompt=prompt, llm=model_llm, memory = memory, output_parser = StrOutputParser())


# Get the user's initial input
text = st.text_input("User: What else would you like to talk about?")

# Run the model on the initial input
if st.button("Submit"):
    output = chain.run(text)
    st.write("PsychologyBot:", output)

while True:
    next_input = st.text_input("User:", key="user_input")

    if should_finish(next_input):
        break

    if st.button("Submit", key="submit_button"):
        output = chain.run(next_input)
        st.write("PsychologyBot:", output)
```
